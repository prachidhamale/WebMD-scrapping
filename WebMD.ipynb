```python
#extracting urls
def get_urls(soups):
    li_urls=[]
    for soup in soups:
        for i in soup.findAll("a",{"data-metrics-link":"link"}):  
            li_urls.append("https://www.webmd.com"+i.get('href'))
    return li_urls
```

```python
soupss=[]
def make_soups(urlss):
    for url in urlss: 
        res=requests.get(url)
        soup= BeautifulSoup(res.text,'lxml')
        soupss.append(soup)
    return soupss 
```

```python
#names
def get_names(soupss):
    li_names=[]
    for soup in soupss:
        for i in soup.findAll("h1"):
            li_names.append(i.text)
    return li_names
```

```python
def get_uses(soupss):
    uses=[]
    for soup in soupss:
        for i in soup.findAll("div",{"id":"tab-1"}):
            s=i.text
            use=s[:s.find(".")]
            uses.append(use)
    return uses
        #print(s)
```

```python
def get_side_effects(soupss):
    side_effects=[]
    for soup in soupss:
        for i in soup.findAll("div",{"id":"tab-2"}):
            s=i.text
            side_effect=s[:s.find(".")]
            side_effects.append(side_effect)
    return side_effects
```

```python
def get_precautions(soupss):
    side_effects=[]
    for soup in soupss:
        for i in soup.findAll("div",{"id":"tab-3"}):
            s=i.text
            precaution=s[:s.find(".")]
            precautions.append(precaution)
    return precautions
```

```python
def get_interactions(soupss):
    interactions=[]
    for soup in soupss:
        for i in soup.findAll("div",{"id":"tab-4"}):
            s=i.text
            interaction=s[:s.find(".")]
            interactions.append(interaction)
    return interactions
```

```python
def get_overdose(soupss):
    overdoses=[]
    for soup in soupss:
        for i in soup.findAll("div",{"id":"tab-5"}):
            s=i.text
            overdose=s[:s.find(".")]
            overdoses.append(overdose)
    return overdoses
```

```python
uses=[]
names=[]
side_effects= []
precautions=[]
interactions=[]
overdose=[]
uses = get_uses(soups)
names = get_names(soups)
side_effects = get_side_effects(soups)
precautions= get_precautions(soups)
interactions=get_interactions(soups)
overdose=get_overdose(soups)
```
